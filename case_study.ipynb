# Broadcasting

```python
df=data['horsepower']=pd.to_numeric(data['horsepower'],errors='coerce')
df_filled=df.fillna(df.mean())
df_filled=df_filled*1.10
df_array=np.array(df_filled)
df_array
```

# Boolean indexing

```python
df =data[data['origin']==2][['displacement']]
df_array=np.array(df)
np.average(df_array)
```

# Matrix Operations

```python
data['horsepower']=pd.to_numeric(data['horsepower'],errors='coerce')
data['horsepower'].fillna(data['horsepower'].mean(),inplace=True)

df_array=np.array(data[['mpg','horsepower','weight']])
v=np.array([1,0.5,-0.2])
df_array.dot(v)
```

# sorting

```python
df_array=np.array(data['model year'])
a=np.argsort(-df_array)
data['car name'].iloc[a[:5]]
```

# Correlation

```python
a=np.array(data['mpg'])
b=np.array(data['weight'])
c=np.corrcoef(a,b)
c[0,1]
```

# Conditional aggregates

```python
a=np.array(data['mpg'])
b=np.array(data['cylinders'])
uniq=np.unique(b)
arr=[]

for i in uniq:
    mean_mpg=a[b==i].mean()
    arr.append((i,mean_mpg))
for i,j in arr:
    print(f"{i}-->{j}")
```

# Pandas

## Basic exploration

```python
data=pd.read_csv('auto-mpg.csv')
data
```

```python
data.head(10)
data.count(axis=0) #rows
data.count(axis=1) #columns
```

```python
data.describe()
```

## Filtering and indexing

```python
data[(data['model year']==75) & (data['weight']<3000)][['car name','weight','mpg']]
```

## Handling Missing Data

```python
data.isnull().sum()
data['horsepower']=pd.to_numeric(data['horsepower'],errors='coerce')
data['horsepower'].fillna(data['horsepower'].mean(),inplace = True)
```

## Data transformation

```python
data['power to weight ratio']=data['horsepower']/data['weight']
data
```

## Group By

```python
a=data.groupby('origin')['mpg'].mean()
a
```

## sort

```python
data.sort_values(by='mpg',ascending=False).head(10)
```

## Apply function

```python
def performance_score(row):
    return row['mpg']*row['acceleration']/row['weight']
data['performance score']=data.apply(performance_score,axis=1)
data
```

## visualization preparation

```python
a=data.groupby('model year')[['mpg','weight','horsepower']].mean()
a
```

## Exporting data

```python
a=data[data['mpg']>30][['mpg','cylinders','horsepower','weight']]
a
a.to_csv('high_mpg_cars.csv')
```

## Finding anomilies

```python
a=np.array(data['mpg'])
q1=np.percentile(a,25)
q3=np.percentile(a,75)
iqr=q3-q1
lb= q1 -(1.5*iqr)
ub=q3+(1.5*iqr)
data[(data['mpg']< lb) | data['mpg']>ub][['car name','mpg','model year']]
```

# Matplotlib


```python
plt.hist(data['mpg'])
plt.xlabel('mpg')
plt.ylabel('count')
plt.show()
```

```python
import seaborn as sns
sns.boxplot(x='cylinders',y='mpg',data=data)
plt.xlabel('number of cylinders')
plt.ylabel('mpg')
plt.show()
```

```python
plt.scatter(data['horsepower'],data['mpg'])
plt.xlabel('horsepower')
plt.ylabel('mpg')
plt.show()
```

```python
sns.regplot(x='weight', y='mpg', data=data, scatter_kws={'alpha':0.6}, line_kws={"color": "red"})
plt.title('Relationship Between Car Weight and MPG')
plt.xlabel('Weight')
plt.ylabel('Miles Per Gallon (MPG)')
```

```python
a=data.groupby('model year')['mpg'].mean()
plt.plot(a.index,a.values)
plt.xlabel('model year')
plt.ylabel('mpg')
plt.show()
```

```python
a=data['origin'].value_counts().sort_index()
plt.bar(a.index,a.values)
plt.xlabel('origin')
plt.ylabel('number of cars')
```

```python
grouped_acc = [data[data['cylinders'] == cyl]['acceleration'] for cyl in sorted(data['cylinders'].unique())]

cylinder_labels = sorted(data['cylinders'].unique())

plt.figure(figsize=(8, 6))
plt.boxplot(grouped_acc, labels=cylinder_labels)

plt.title('Acceleration Distribution by Number of Cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel('Acceleration')
```

```python
car_counts_by_year = data['model year'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
plt.bar(car_counts_by_year.index, car_counts_by_year.values, color='lightcoral')
plt.xlabel('Model Year')
plt.ylabel('Number of Cars')
plt.title('Number of Car Entries by Model Year')
plt.grid(axis='y')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(data['weight'], data['horsepower'], data['mpg'], c='teal', alpha=0.7)

ax.set_xlabel('Weight')
ax.set_ylabel('Horsepower')
ax.set_zlabel('MPG')
ax.set_title('3D Scatter Plot of Weight, Horsepower, and MPG')

plt.show()
```

```python
top_10 = data.sort_values(by='mpg', ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.barh(top_10['car name'], top_10['mpg'], color='mediumseagreen')
plt.xlabel('MPG')
plt.title('Top 10 Fuel Efficient Cars')
plt.gca().invert_yaxis()
plt.show()

```
